```python
import pandas as pd
df=pd.read_csv('/content/mtcars.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
x=df['mpg']
y=df['wt']
plt.bar(x,y)
plt.xlabel=('miles per gallon')
plt.ylabel=('frequency')
plt.title=('bar')
plt.grid=('True')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.hist(df['mpg'],bins=10,color='red',edgecolor='black')
plt.xlabel=('weight')
plt.ylabel=('miles per Gallon')
plt.title=('histogram')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('/content/mtcars.csv')
plt.boxplot(df['mpg'])
plt.xlabel=('weight')
plt.ylabel=('miles per Gallon')
plt.title=('histogram')
plt.grid=('true')
plt.show()
```
