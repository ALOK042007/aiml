```python
# 1️⃣ Import Libraries
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import tensorflow_datasets as tfds

# 2️⃣ Load Dataset (TF Flowers)
dataset, info = tfds.load('tf_flowers', with_info=True, as_supervised=True)

# Split dataset (80% train, 20% test)
train_size = 0.8
train_count = int(info.splits['train'].num_examples * train_size)
test_count = info.splits['train'].num_examples - train_count

train_data = dataset['train'].take(train_count)
test_data = dataset['train'].skip(train_count)

# 3️⃣ Preprocessing (resize + normalize)
IMG_SIZE = 180

def format_image(image, label):
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
    image = image / 255.0
    return image, label

train_data = train_data.map(format_image).batch(32).shuffle(1000)
test_data = test_data.map(format_image).batch(32)

# 4️⃣ Visualize sample images
class_names = info.features['label'].names

plt.figure(figsize=(8,8))
for images, labels in train_data.take(1):
    for i in range(9):
        plt.subplot(3,3,i+1)
        plt.imshow(images[i])
        plt.title(class_names[labels[i].numpy()])
        plt.axis('off')
plt.tight_layout()
plt.show()

# 5️⃣ Build CNN Model
cnn_model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 3)),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(len(class_names), activation='softmax')
])

# 6️⃣ Compile CNN Model
cnn_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
cnn_model.summary()

# 7️⃣ Train CNN Model
cnn_history = cnn_model.fit(train_data, epochs=5, validation_data=test_data)

# 8️⃣ Evaluate CNN Model
cnn_test_loss, cnn_test_acc = cnn_model.evaluate(test_data)
print(f"✅ CNN Test Accuracy: {cnn_test_acc*100:.2f}%")

# 9️⃣ Plot CNN Accuracy & Loss
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(cnn_history.history['accuracy'], label='Train Acc')
plt.plot(cnn_history.history['val_accuracy'], label='Val Acc')
plt.legend()
plt.title("CNN Accuracy Curve")

plt.subplot(1,2,2)
plt.plot(cnn_history.history['loss'], label='Train Loss')
plt.plot(cnn_history.history['val_loss'], label='Val Loss')
plt.legend()
plt.title("CNN Loss Curve")
plt.show()

# 🔟 Confusion Matrix & Classification Report (CNN)
y_true = []
y_pred = []

for images, labels in test_data:
    preds = cnn_model.predict(images)
    y_true.extend(labels.numpy())
    y_pred.extend(np.argmax(preds, axis=1))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, cmap="Purples", xticklabels=class_names, yticklabels=class_names)
plt.title("CNN Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

print("\n📊 CNN Classification Report:\n", classification_report(y_true, y_pred, target_names=class_names))

# ===============================================
# ⚡️ NEW: Linear Model (Baseline)
# ===============================================

linear_model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(IMG_SIZE, IMG_SIZE, 3)),  # Flatten image pixels
    tf.keras.layers.Dense(len(class_names), activation='softmax')  # Direct classification
])

linear_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
linear_model.summary()

# Train Linear Model
linear_history = linear_model.fit(train_data, epochs=5, validation_data=test_data)

# Evaluate Linear Model
linear_test_loss, linear_test_acc = linear_model.evaluate(test_data)
print(f"✅ Linear Model Test Accuracy: {linear_test_acc*100:.2f}%")

# Compare CNN vs Linear Accuracy
plt.figure(figsize=(8,5))
plt.bar(['Linear Model', 'CNN Model'], [linear_test_acc*100, cnn_test_acc*100], color=['orange', 'green'])
plt.ylabel('Accuracy (%)')
plt.title('Model Performance Comparison')
plt.show()
```
